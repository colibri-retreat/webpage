```python
import math
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.patches import FancyArrowPatch
```

```python
fig, ax = plt.subplots()

# Number of notes
n_points = 12

# Angles for each point (in radians), adjusted to start at the top
angles = np.linspace(2 * np.pi, 0, n_points, endpoint=False) + (np.pi / 2)
x_coords = np.cos(angles)
y_coords = np.sin(angles)

# Tonal Names
notes_12_tet = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]

# Plot the points with numbers
for i in range(n_points):
    sqrt_text = '$\sqrt[12]{2}$'
    ax.text(1.3 * x_coords[i], 1.3 * y_coords[i], notes_12_tet[i], ha='center', va='center', size=20)
    ax.plot([0.95 * x_coords[i], 1.05 * x_coords[i]], [0.95 * y_coords[i], 1.05 * y_coords[i]], color='black')

# Draw lines connecting each point to the next
for i in range(n_points):
    ax.plot([x_coords[i], x_coords[(i + 1) % n_points]], [y_coords[i], y_coords[(i + 1) % n_points]], color='r', linestyle='solid', linewidth=2)

# Add curved arrows from one radial line to the next and labels
for i in range(n_points):
    start = (x_coords[i], y_coords[i])
    end = (x_coords[(i + 1) % n_points], y_coords[(i + 1) % n_points])
    arrow = FancyArrowPatch(start, end, connectionstyle="arc3, rad=0.4", arrowstyle='->', color='b', linewidth=2, shrinkA=10, shrinkB=10, mutation_scale=10)
    ax.add_patch(arrow)

ax.set_aspect('equal')
ax.axis('off')
```

```python
fig, ax = plt.subplots()

# Number of notes
n_points = 12

# Angles for each point (in radians), adjusted to start at the top
angles = np.linspace(2 * np.pi, 0, n_points, endpoint=False) + (np.pi / 2)
x_coords = np.cos(angles)
y_coords = np.sin(angles)

# Tone names
notes_12_tet = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
notes_19_tet = ["A", "A#", "Bb", "B", "B#", "Cb", "C", "C#", "Db", "D", "D#", "Eb", "E", "E#", "Fb", "F", "F#", "Gb", "G", "G#", "Ab", "A"]

# Plot the points with numbers
for i in range(n_points):

    # 12 tet system
    if n_points == 12:
        sqrt_text = '$\sqrt[12]{2}$'
        ax.text(1.2 * x_coords[i], 1.2 * y_coords[i], notes_12_tet[i], ha='center', va='center', size=15)

    # 19 tet system
    elif n_points == 19:
        sqrt_text = '$\sqrt[19]{2}$'
        ax.text(1.2 * x_coords[i], 1.2 * y_coords[i], notes_19_tet[i], ha='center', va='center', size=15)

    # All other systems
    else:
        sqrt_text = ''
        ax.text(1.2 * x_coords[i], 1.2 * y_coords[i], str(i + 1), ha='center', va='center', size=15)

    # Radial lines
    ax.plot([0.95 * x_coords[i], 1.05 * x_coords[i]], [0.95 * y_coords[i], 1.05 * y_coords[i]], color='black')

# Draw lines connecting each point to the next
for i in range(n_points):
    ax.plot([x_coords[i], x_coords[(i + 1) % n_points]], [y_coords[i], y_coords[(i + 1) % n_points]], color='r', linestyle='solid')

# Add curved arrows from one radial line to the next and labels
for i in range(n_points):
    start = (x_coords[i], y_coords[i])
    end = (x_coords[(i + 1) % n_points], y_coords[(i + 1) % n_points])
    arrow = FancyArrowPatch(start, end, connectionstyle="arc3, rad=0.4", arrowstyle='->', color='b', linewidth=1.5, shrinkA=10, shrinkB=10, mutation_scale=10)
    ax.add_patch(arrow)

    # Calculate the midpoint for the text
    mid_x = (start[0] + end[0]) / 2 * 0.75
    mid_y = (start[1] + end[1]) / 2 * 0.75
    ax.text(mid_x, mid_y, sqrt_text, ha='center', va='center', size=10, color='b')

ax.set_aspect('equal')
ax.axis('off')
```

```python
import numpy as np

# Calculate the percentage error
def calculate_error(exact, approx):
    return abs((exact - approx) / exact) * 100

# Rational approximation
def find_best_rational_approx(ratio):
    best_approx = None
    smallest_error = float('inf')
    best_fraction = (1, 1)

    for numerator in range(1, 21):
        for denominator in range(1, 21):
            approx = numerator / denominator
            error = calculate_error(ratio, approx)
            if error < smallest_error:
                smallest_error = error
                best_approx = approx
                best_fraction = (numerator, denominator)

    return best_fraction, smallest_error

def calculate_tet_approximations(tet_system):
    total_error = 0
    print(f"\nApproximations for {tet_system}-TET System:")
    for n in range(tet_system + 1):
        ratio = 2 ** (n / tet_system)  # Exact ratio in the TET system
        best_fraction, error_percent = find_best_rational_approx(ratio)  # Find best approximation

        total_error += error_percent  # Accumulate the error
        print(f"Note {n}: Ratio = {ratio:.6f}, Best Approx = {best_fraction[0]}/{best_fraction[1]}, Error = {error_percent:.6f}%")

    average_error = total_error / tet_system  # Calculate the average error
    print(f"\nTotal Error for {tet_system}-TET System: {total_error:.6f}%")
    print(f"Average Error per Note: {average_error:.6f}%")

# Calculate for 12-TET and 19-TET systems
calculate_tet_approximations(12)
calculate_tet_approximations(19)
```

```python
# Function to calculate the percentage error
def calculate_error(exact, approx):
    return abs((exact - approx) / exact) * 100

# Function to find the best rational approximation with numerators and denominators from a specified range
def find_best_rational_approx(ratio, num_range):
    best_approx = None
    smallest_error = float('inf')
    best_fraction = (1, 1)

    for numerator in range(1, num_range + 1):
        for denominator in range(1, num_range + 1):
            approx = numerator / denominator
            error = calculate_error(ratio, approx)
            if error < smallest_error:
                smallest_error = error
                best_approx = approx
                best_fraction = (numerator, denominator)

    return smallest_error

# Function to calculate the average and maximum error for a given TET system with a specified fraction range
def calculate_errors(tet_system, num_range):
    total_error = 0
    max_error = 0

    for n in range(tet_system):
        ratio = 2 ** (n / tet_system)  # Exact ratio in the TET system
        error_percent = find_best_rational_approx(ratio, num_range)  # Find best approximation error
        total_error += error_percent  # Accumulate the error
        if error_percent > max_error:
            max_error = error_percent  # Update the maximum error

    average_error = total_error / tet_system  # Calculate average error
    return average_error, max_error

# Function to calculate errors for TET systems from 1 to a specified number and plot the results
def plot_tet_errors_with_melodies(max_tet, num_range):
    tet_range = range(1, max_tet + 1)
    average_errors = []
    max_errors = []
    num_melodies = []

    for tet in tet_range:
        avg_error, max_error = calculate_errors(tet, num_range)
        average_errors.append(avg_error)
        max_errors.append(max_error)
        num_melodies.append(tet ** 10)  # Number of possible 10-tone melodies

    # Plotting the average and maximum errors with the number of melodies on a secondary axis
    fig, ax1 = plt.subplots(figsize=(7, 4))

    x = tet_range
    y = max_errors
    p = np.poly1d(np.polyfit(x, y, 3))

    # ax1.plot(tet_range, max_errors, marker='o', linestyle='-', color='r', label='Maximum Error')
    ax1.plot(x, p(x), linestyle='-', color='r', label='Maximum Error')
    ax1.set_xlabel('TET System', size=14)
    ax1.set_ylabel('Maximum Error (%)', color='r', size=14)
    ax1.set_title(f'Tonal Variety versus Tonal Complexity\n', size=14)
    ax1.tick_params(axis='y', labelcolor='r')
    ax1.set_xticks(tet_range[::2])

    # Creating a secondary y-axis for the number of melodies with a logarithmic scale
    ax2 = ax1.twinx()
    ax2.plot(tet_range, num_melodies, linestyle='-', color='b', label='Number of Melodies')
    ax2.set_ylabel('Number of\n10-tone Melodies (log)', color='b', size=14)
    ax2.set_yscale('log')
    ax2.tick_params(axis='y', labelcolor='b')

    # Legends
    fig.tight_layout()  # To adjust the layout
    plt.show()

# Example usage: Plotting for TET systems from 1 to 27 with fractions from 1 to 20
plot_tet_errors_with_melodies(27, 20)
```
