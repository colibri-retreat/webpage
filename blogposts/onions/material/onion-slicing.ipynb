```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 1000)

plt.figure(figsize=(4, 4))

for i in np.linspace(1, 8, 10):
  plt.plot(x, x**i, c='k', lw=(i/5))
  plt.plot(x, x**(1/i), c='k', lw=(i/5))

plt.axis('off')
plt.savefig('thumbnail.png', dpi=200)
```

```python
# Define the radii of the half circles
radii = np.linspace(1, 0.1, 7)

# Create a figure and axis
fig, ax = plt.subplots()

# Draw the upper half circles
for radius in radii:
    # Define the x-coordinates for the semicircle
    x = np.linspace(-radius, radius, 1000)

    # Define the corresponding y-coordinates for the semicircle
    y = np.sqrt(radius**2 - x**2)

    # Plot the semicircle
    ax.plot(x, y, color='k')

# Set the aspect ratio to 'equal' to maintain circular shape
ax.set_aspect('equal')

# Set the x and y axis limits
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(0, 1.5)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Onion slice view\n', size=14)

# Show the plot
plt.savefig('onion-slices.png', dpi=200)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the radii of the half circles
radii = np.linspace(1, 0.1, 7)

# Create a figure and axis
fig, ax = plt.subplots()

# Draw the upper half circles
for radius in radii:
    # Define the x-coordinates for the semicircle
    x = np.linspace(-radius, radius, 1000)

    # Define the corresponding y-coordinates for the semicircle
    y = np.sqrt(radius**2 - x**2)

    # Plot the semicircle
    ax.plot(x, y, color='k')

# Define radial cuts
cuts = np.linspace(0, np.pi/2, 7)

# Draw radial cuts
for cut in cuts[1:]:
    x = np.cos(cut) * 1.2
    y = np.sin(cut) * 1.2
    ax.plot([0, x], [-0.5573066929856644, y], color='k', ls='dashed', lw=1, clip_on=False)

# Set the aspect ratio to 'equal' to maintain circular shape
ax.set_aspect('equal')

# Set the x and y axis limits
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(0, 1.5)

# Set labels and title
ax.set_ylabel('Y')
ax.set_title('Radial cutting aimed below center\n', size=14)

# Show the plot
plt.savefig('optimal-cuts.png', dpi=200)
```
